```python
#address to coordinates = geocoding

from geopy.geocoders import ArcGIS
nom = ArcGIS()
n = nom.geocode("2618 T. Earnshaw St., Gagalangin, Tondo, Manila")
```

```python
df5
```

```python
df5["Address"] = df5["Address"] + "," + df5["City"] + "," + df5["State"] + "," + df5["Country"]
```

```python
df5
```

```python
# no need to iterate in.. pandas #what method you want to .apply to the column
df5["Coordinates"] = df5["Address"].apply(nom.geocode)
```

```python
df5
```

```python
#lambda in  single line function acting like a function
df5["Latitude"] = df5["Coordinates"].apply(lambda x: x.latitude if x!= None else None)
df5["Longitude"] = df5["Coordinates"].apply(lambda x: x.longitude if x!= None else None)
df5
```

```python
df5
```

```python
lat = list(df5["Latitude"])
long = list(df5["Longitude"])
zipl = zip(lat,long)
for i,coord in enumerate(zipl):
    for x in coord:
        print(x)
```
